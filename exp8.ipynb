```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_excel("student.xlsx")
x = np.array(df["name"])
y = np.array(df["mark"])
plt.tick_params(axis="x",labelrotation=90)
plt.xlabel("students")
plt.ylabel("marks")
plt.bar(x,y)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_excel("student.xlsx")
y = np.array(df["mark"])
category=[0,0,0,0,0]
for i in y:
    if(i<=50):
        category[0]=category[0]+1
    elif(i>50 and i<=60):
        category[1]=category[1]+1
    elif(i>60 and i<=70):
        category[2]=category[2]+1
    elif(i>70 and i<=80):
        category[3]=category[3]+1
    elif(i>80 and i<=100):
        category[4]=category[4]+1
label =["fail","third class","second class","first class","distinction"]
plt.pie(category,labels=label)

plt.show()
```
