```python
import matplotlib.pyplot as plt
import numpy as np

#If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3
plt.plot([1, 2, 3, 4])
plt.ylabel('y label')
plt.xlabel('x label')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot
x = [1,2]
y = [2,3]
plt.plot(x,y)
plt.show()
```

```python
#only points

x = np.array([1,2])
y = np.array([2,3])
plt.plot(x,y,'o')
plt.show()
```

```python
#multiple points

x= np.array([1, 2, 3, 4])
y = np.array([1, 2, 9, 6])
plt.plot(x,y)
plt.show()
```

```python
#marke the points and plot

x= np.array([1, 2, 3, 4])
y = np.array([1, 2, 9, 6])
plt.plot(x,y,marker='o')
plt.show()
```

```python
#shortcut 'typeofpoint|typeofline|color'
#markersize ms='x'
#marker color mfc = 'r'

x= np.array([1, 2, 3, 4])
y = np.array([1, 2, 9, 6])
plt.plot(x,y,'o:r',ms='20',mfc='g')   
plt.show()
```

```python
#plot multiple graphs

x= np.array([1, 2, 3, 4])
y = np.array([1, 2, 9, 6])

x1= np.array([0, -2, -3, 7])
y1 = np.array([4, 2, 0, 6])


plt.plot(x,y,":g")   
plt.plot(x1,y1,"-r")   
plt.show()
```

```python
#labels for plot
x1= np.array([0, -2, -3, 7])
y1 = np.array([4, 2, 0, 6])

plt.plot(x1,y1,"-r")   

plt.xlabel("salary")
plt.ylabel("age")
plt.title("my graph")
plt.show()
```

```python
#add gridlines to graph
x= np.array([1, 2, 3, 4])
y = np.array([1, 2, 9, 6])


plt.plot(x,y,":r")   
plt.grid()
plt.show()
```

```python
#subplots (#rows,#cols,#serialno)

x= np.array([1, 2, 3, 4])
y = np.array([1, 2, 9, 6])

plt.subplot(2,1,1)
plt.plot(x,y,":g") 


x1= np.array([0, -2, -3, 7])
y1 = np.array([4, 2, 0, 6])


plt.subplot(2,1,2) 
plt.plot(x1,y1,"-r")   
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.subplot(2,1,1) 
plt.bar(x,y)

plt.subplot(2,1,2) 
plt.barh(x,y)
plt.show()
```

```python
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

print("mean = ",np.mean(speed))
print("median = ",np.median(speed))
print("mode = ",stats.mode(speed))
print("percentile of 86 = ",np.percentile(speed,100))

```
